```
import pandas as pd
```

loading the dataset

```
data = pd.read_csv('./test.csv', encoding='ISO-8859-1')
data
```

cleaning the na valuesa dnd datas

```
print(data.describe() ,data.isnull().sum())
```

```

data = data.dropna()
data.head(10)
```

label encoding the target

```
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data["sentiment"] = encoder.fit_transform(data["sentiment"])
data
```

loading the input and target

```
import re

def clean_text(text):
    text = re.sub(r"http\S+", "", text)  # Remove URLs
    text = re.sub(r"[^a-zA-Z\s]", "", text)  # Remove special characters
    text = text.lower()  # Convert to lowercase
    text = re.sub(r"\s+", " ", text).strip()  # Remove extra spaces
    return text

data['text'] = data['text'].apply(clean_text)
data
```

```
X_datas = data['text']
y_datas = data['sentiment']
```

spliting the training set

```
from sklearn.model_selection import train_test_split
X_data,X_test,y_data,y_test = train_test_split(X_datas,y_datas,test_size=0.2,random_state=42)
```

loading the models

```
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
```

FITTING THE MODEL

```
model = make_pipeline(TfidfVectorizer(max_features=5000, ngram_range=(1, 2), stop_words='english'), MultinomialNB())
model.fit(X_data,y_data)
```

```
y_pred = model.predict(X_test)
```

```
from sklearn.metrics import classification_report, accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy for  Mulyinominal : {accuracy}")

```
