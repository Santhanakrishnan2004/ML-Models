MultinominalNB with tfdf for SMS spam prediction , dataset : https://www.kaggle.com/datasets/uciml/sms-spam-collection-dataset/data

```
#importing the required fucntions

from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import TfidfVectorizer # it is used to chenage the text to numericals
from sklearn.naive_bayes import MultinomialNB               #classification model based on Multinomial Naive Bayes
from sklearn.model_selection import train_test_split        # used to split the data serts into train and test
from sklearn.metrics import classification_report, accuracy_score
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```
# Load the dataset
data = pd.read_csv("/spam.csv", encoding="latin-1")
data = data.rename(columns={"v1": "label", "v2": "message"})
data
```

```

# Encode target labels changing spam and ham to 0 and 1 numericals
label_encoder = LabelEncoder()
data["label"] = label_encoder.fit_transform(data["label"])

data
```

```
X_train, X_test, y_train, y_test = train_test_split(data["message"], data["label"], test_size=0.2, random_state=42)
```

```
model = make_pipeline(TfidfVectorizer(),MultinomialNB()) # initiating pipleine with the model and vectorization
model.fit(X_train,y_train)              #fitting the training data to the model
result = model.predict(X_test)           #prediciting the model with the test set
```

```
# Evaluation
print("Accuracy:", accuracy_score(y_test, result))
print("\nClassification Report:\n", classification_report(y_test, result))
```
