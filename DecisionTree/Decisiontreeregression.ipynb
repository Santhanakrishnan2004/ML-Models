```
from sklearn.datasets import load_diabetes
datasets=load_diabetes()
```

```
print(datasets['DESCR'])
```

```
import pandas as pd
#indepe variable
dfdiab=pd.DataFrame(datasets['data'],columns=datasets['feature_names'])
```

```
dfdiab
```

```
X=dfdiab
y=datasets['target']
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
```

```
X_train.corr()
```

```
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(X_train,y_train)
```

```
y_pred=regressor.predict(X_test)
```

```

from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)



from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Absolute Error: {mae}")
print(f"R-squared: {r2}")
print(f"Mean Squared Error: {mse}")
```

```
param={
    'criterion':['squared_error','friedman_mse','absolute_error'],
    'splitter':['best','random'],
    'max_depth':[1,3,5,7,9,11,12,15,20,25],
    'max_features':['auto','sqrt','log2']

}
```

```
from sklearn.model_selection import GridSearchCV
regressors=DecisionTreeRegressor()
grid= GridSearchCV(regressors , param_grid=param,scoring='neg_mean_squared_error',cv=5)
```

```
grid.fit(X_train,y_train)
```

```
grid.best_params_
```

```
y_pred=grid.predict(X_test)
```

```


from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Absolute Error: {mae}")
print(f"R-squared: {r2}")
print(f"Mean Squared Error: {mse}")
```

```
trees=DecisionTreeRegressor(criterion='squared_error',max_depth=3,max_features='sqrt',splitter='best')
trees.fit(X_train,y_train)
```

```
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,20))
tree.plot_tree(trees,filled=True)
```
